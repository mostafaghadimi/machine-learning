## حذف کردن داده
</div>

<div dir="rtl">
در روش اول که «حذف کردن داده‌های گم‌شده» است، همان‌طور که از نامش انتظار می‌رود، ما شروع به حذف داده‌های گم‌شده (از دست رفته) می‌کنیم، این کار را همان‌طور که پیش‌تر دیده شد، با استفاده از متد dropna انجام می‌دهیم. اگر شرط خاصی را در آن اثر ندهیم، تمام ردیف‌هایی را که مقدار nan دارند، از dataframe فعلی حذف کرده و یک dataframe جدید برمی‌گرداند.
 اگر بخواهیم در dataframe فعلی هم اعمال شود باید آرگومان inplace=True را به تابع پاس دهیم. به روش توضیح داده شده روش Listwise گفته می‌شود. در صورتی که داده‌هایی که بعضی از آن‌ها از دست رفته‌اند در تحلیل استفاده شود به آن Pairwise گفته می‌شود که شامل مشکل‌هایی هم نیز هست که توضیح آن در این‌جا آورده شده است.
در حالت آخر اگر در بیش از حدود ۶۰ درصد مشاهدات، داده‌ها از دست رفته باشند، عقلانی‌تر است که آن متغیر را حذف کنیم (اگر خیلی اثرگذار نباشد.) 
</div>

##  جای‌گزینی داده
</div>

<div dir="rtl">
در روش جای‌گزینی ما سعی می‌کنیم تا حدس‌های معقولی بزای داده‌های از دست رفته بزنیم. البته این روش وقتی کارساز است که درصد داده‌های گم‌شده نسبت به کل داده‌ها درصد کمی را گرفته باشد. رایج‌ترین کارها جای‌گزین کردن با میانگین یا میانه است.. البته این روش‌ها خود مشکلاتی دارند که یکی از آن‌ها از بین رفتن تنوع است در صورتی که تعداد داده‌های گم‌شده زیاد باشد. 
</div>

```python
# mean example
df.fillna(df.mean())
```

```python
# Replace with most frequent value of each column
df.apply(lambda col: col.fillna(col.value_counts().index[0])) #first solution!
```

```python
# Replace with most frequent value of each column 
df.fillna(df.mode().iloc[0]) #second solution
```

```python
#median example
df.fillna(df.median())
```

```python
# interpolation example
df.interpolate(method ='linear', limit_direction ='forward') 
```

<div dir="rtl">
تا اینجای کار فقط داشتیم داده‌هامون رو تمیز می‌کردیم. از اینجا به بعد می‌خوایم کمی روی اونها کار کنیم. یکی از 
فاکتورهای مهم توی کار کردن میزان حقوقه. پس بیاید و ستون تخمین حقوق رو به دو ستون تبدیل کنید که حداقل و حداکثر حقوق 
درخواستی در اون درج شده.
سپس ستون میانگین حقوق را از این دو ستون جدید به دست بیارید و در انتها پنج داده‌ی اول مجموعه داده رو نمایش بدید.
    (۸ امتیاز)
</div>

```python
import re
```

```python
def min_max_average(expected):
    minimum, maximum = re.findall(r"\$\d+K", expected)
    minimum = int(minimum[1:-1])
    maximum = int(maximum[1:-1])
    average = (minimum + maximum) / 2
    return minimum, maximum, average

df["Minimum"], df["Maximum"], df["Average"] = zip(*df["Salary Estimate"].map(min_max_average))
df.head()
```

```python
df.shape
```

<div dir="rtl">
    می‌خوایم یه آمار کلی از حقوق کارها داشته باشیم. کاری که می‌خوایم بکنید اینه که ۵٪ از کارها با بالاترین میانگین حقوق و ۵٪ از کارها با پایین‌ترین میانگین حقوق رو حذف کنید و بعدش یه آمار کلی از میانگین حقوق‌ها بدید و همچنین بگید که بنظرتون چرا ۵٪ از بالا و ۵٪ از پایین رو حذف کردیم؟ آمار کلی‌ای که می‌خوایم شامل:  mean, std, min, max می‌شه.
    (۶ امتیاز)
</div>
