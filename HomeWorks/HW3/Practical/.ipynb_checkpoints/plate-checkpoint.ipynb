<div dir="rtl">
    <h3>
        تشکیل ماتریس phee و محاسبه W بهینه
    </h3>
</div>
<p></p>


```python
target = train_set['MEDV']
phee_df = train_set.drop(['MEDV'], axis=1)
phee_df['bayas'] = 1
phee_df
```

```python
import numpy as np

phee = phee_df.to_numpy()
phee_t = phee.transpose()
phee_t_mul_phee = np.matmul(phee_t, phee)
inverse = np.linalg.inv(phee_t_mul_phee)
w = np.matmul(np.matmul(inverse, phee_t), target)
w
```

<div dir="rtl">
   با محاسبه moore penrose pseudo inverse مقدار بهینه برای پارامترها حساب شد.
<p></p>
بنابراین مقدار تارگت برای هر داده از ضرب w_transpose در بردار ویژگی‌های آن داده حساب می‌شود.    
</div>


```python
def predict(row, w):
    x = row.to_numpy().transpose()
    return np.matmul(w.transpose(), x)

train_features = train_set.drop(['MEDV'], axis = 1)
train_features['bayas'] = 1
train_target = train_set['MEDV']
train_predicted = train_features.apply(lambda row: predict(row, w), axis=1)

test_set = pd.read_csv('./test_set.csv')
test_features = test_set.drop(['MEDV'], axis = 1)
test_features['bayas'] = 1
test_target = test_set['MEDV']
test_predicted = test_features.apply(lambda row: predict(row, w), axis=1)
```

<div dir="rtl">
   در قسمت بالا تارگت برای داده‌های یادگیری و تست محاسبه شد. در ادامه با رسم نمودار به تحلیل آن‌ها با مقادیر واقعی‌شان پرداخته می‌شود. 
</div>

```python
def second_order(column, df):
    name = column.name + "2"
    df[name] = column * column

phee2_df = train_set.drop(['MEDV'], axis=1)
phee2_df.apply(lambda col: second_order(col, phee2_df), axis = 0)
phee2_df['bayas'] = 1
phee2_df
```

    <h4>
        ۱۰ ویژگی جدید را به ویژگی‌های اصلی اضاقه میکنیم.
    </h4>
</div>
<p></p>


```python
def get_gaussian_basis(x, mue):
    return np.exp(-((np.linalg.norm(np.subtract(x, mue), ord=2) ** 2) / 2))

def set_gaussian_features(mues, df):
    source_df = df.copy()
    for i, mue in enumerate(mues):
        df['gauss' + str(i)] = source_df.apply(lambda row: get_gaussian_basis(row.to_numpy(), mue), axis=1)
    

mues = train_set.sample(10).drop(['MEDV'], axis=1).to_numpy()
phee3_df = train_set.drop(['MEDV'], axis=1)
set_gaussian_features(mues, phee3_df)
phee3_df['bayas'] = 1
phee3_df
```

```python
phee3 = phee3_df.to_numpy()
phee3_t = phee3.transpose()
phee3_t_mul_phee3 = np.matmul(phee3_t, phee3)
inverse3 = np.linalg.pinv(phee3_t_mul_phee3)
w3 = np.matmul(np.matmul(inverse3, phee3_t), target)
w2
```

<div dir="rtl">
   با محاسبه moore penrose pseudo inverse مقدار بهینه برای پارامترها حساب شد.
<p></p>
بنابراین مقدار تارگت برای هر داده از ضرب w3_transpose در بردار ویژگی‌های آن داده حساب می‌شود.    
</div>


```python
train_features3 = phee3_df.copy()
train_predicted3 = train_features3.apply(lambda row: predict(row, w3), axis=1)

test_features3 = test_set.drop(['MEDV'], axis = 1)
set_gaussian_features(mues, test_features3)
test_features3['bayas'] = 1
test_predicted3 = test_features3.apply(lambda row: predict(row, w3), axis=1)
```

<div dir="rtl">
    <h4>
        مقدار MSE از رابطه زیر به دست می‌آید.
    </h4>
</div>


```python
def get_MSE(real_data, predicted_data):
    return (np.linalg.norm(np.subtract(real_data, predicted_data), ord=2) ** 2) / real_data.size
```

<div dir="rtl">
    <h4>
        مقادیر MSE خواسته شده با در نظر گرفتن ۳ دسته ویژگی روی داده‌های تست و یادگیری به شرح زیر است.
    </h4>
</div>


```python
train_main = get_MSE(train_target, train_predicted)
test_main = get_MSE(test_target, test_predicted)
```

```python
train_2nd = get_MSE(train_target, train_predicted2)
test_2nd = get_MSE(test_target, test_predicted2)
```

```python
train_gauss = get_MSE(train_target, train_predicted3)
test_gauss = get_MSE(test_target, test_predicted3)
```
