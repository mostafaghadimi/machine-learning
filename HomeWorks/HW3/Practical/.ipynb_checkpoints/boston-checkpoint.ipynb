## Loading data

```python
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['MEDV'] = boston.target
```

```python
for column in X_train.columns:
    plt.scatter(X_train[column], y_train)
    plt.xlabel(column)
    plt.ylabel('MEDV')
    plt.show()
```

```python
X_tr = np.array(X_train)
y_train = np.array(y_train)
```

```python
w = np.linalg.inv(X_tr.T.dot(X_tr)).dot(X_tr.T).dot(y_train)
w.shape
```

```python
y_train_predicted = X_tr.dot(w)
```

```python
plt.scatter(X_train['LSTAT'], y_train_predicted, label="predicted")
plt.scatter(X_train['LSTAT'], y_train, label="real")
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.legend()
plt.show()
```

```python
plt.scatter(X_train['DIS'], y_train_predicted, label="predicted")
plt.scatter(X_train['DIS'], y_train, label="real")
plt.xlabel('DIS')
plt.ylabel('MEDV')
plt.legend()
plt.show()
```

## Second Order

```python
X_tr_second = np.append(X_tr, np.square(X_tr), axis=1)
```

```python
w_second = np.linalg.inv(X_tr_second.T.dot(X_tr_second)).dot(X_tr_second.T).dot(y_train)
w.shape
```

```python
y_train_second_predicted = X_tr_second.dot(w_second)
```

```python
plt.scatter(X_train['LSTAT'], y_train_second_predicted, label="predicted")
plt.scatter(X_train['LSTAT'], y_train, label="real")
plt.legend()
plt.xlabel("LSTAT")
plt.show()
```

```python
plt.scatter(X_train['DIS'], y_train_second_predicted, label="predicted")
plt.scatter(X_train['DIS'], y_train, label="real")
plt.legend()
plt.xlabel("DIS")
plt.show()
```
